```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('data/experimental/edmund_71Gpa_run1.csv')
df2 = pd.read_csv('data/experimental/edmund_71Gpa_run2.csv')
df1['temp normed'] = (df1['temp'] - df1['temp'].iloc[0]) / (df1['temp'].max() - df1['temp'].min())
df2['temp normed'] = (df2['temp'] - df2['temp'].iloc[0]) / (df2['temp'].max() - df2['temp'].min())
df1['oside normed'] = (df1['oside'] - df1['oside'].iloc[0]) / (df1['temp'].max() - df1['temp'].min())
df2['oside normed'] = (df2['oside'] - df2['oside'].iloc[0]) / (df2['temp'].max() - df2['temp'].min())
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(df1['time'], df1['temp normed'], label='Run 1 pside')
ax[0].plot(df2['time'], df2['temp normed'], label='Run 2 pside')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Temperature')
ax[0].set_title('Temperature vs Time for pside')
ax[0].legend()
ax[1].plot(df1['time'], df1['oside normed'], label='Run 1 oside')
ax[1].plot(df2['time'], df2['oside normed'], label='Run 2 oside')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Temperature')
ax[1].set_title('Temperature vs Time for oside')
ax[1].legend()
plt.tight_layout()
plt.show()
```
